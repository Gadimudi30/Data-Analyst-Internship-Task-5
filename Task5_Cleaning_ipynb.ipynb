```python
import pandas as pd

df = pd.read_csv("tested.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Fare"] = df["Fare"].fillna(df["Fare"].median())
```

```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
df.drop(columns=["Cabin"], inplace=True)
```

```python
df.drop_duplicates()
```

```python
df["Survived"] = df["Survived"].astype(int)
```

```python
df["Age_Group"] = pd.cut(
    df["Age"],
    bins=[0, 12, 19, 35, 60, 100],
    labels=["Child", "Teen", "Young Adult", "Adult", "Senior"]
)
```

```python
df.to_csv("cleaned_data.csv", index=False)
```

In this task, I worked with the Titanic dataset using Python and Pandas.

First, I loaded the dataset using the read_csv() function and checked the first few rows to understand the data.

Then, I used info() and isnull().sum() to identify missing values and understand the structure of the dataset.

I found missing values in columns like Age, Fare, and Cabin. For numeric columns such as Age and Fare, I filled the missing values using the median to avoid the effect of outliers.

For the Embarked column, which is categorical, I filled missing values using the most frequent value (mode).

The Cabin column had too many missing values, so I removed it as it was not useful for analysis.

After cleaning missing values, I checked for duplicate records and removed them to ensure data quality.

I also converted the Survived column to the correct datatype for accurate analysis.

To show data transformation skills, I created a new column called Age_Group by categorizing passengers based on their age.

Finally, I saved the cleaned dataset as cleaned_data.csv for further analysis and reporting.
